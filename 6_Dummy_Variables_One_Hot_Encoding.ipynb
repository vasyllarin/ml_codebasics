```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('homeprices1.csv')
df
```

```python
dummies = pd.get_dummies(df['town'], dtype=int)
dummies
```

```python
merged = pd.concat([df, dummies], axis=1)
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis=1)
final
```

```python
X = final.drop(['price'], axis=1)
X
```

```python
y = final['price']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X, y)
```
